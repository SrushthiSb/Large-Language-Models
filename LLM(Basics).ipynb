```python
from transformers import AutoModelForCausalLM, AutoTokenizer
llm=AutoModelForCausalLM.from_pretrained('facebook/opt-125m')
tok=AutoTokenizer.from_pretrained('facebook/opt-125m')
```

```python
model_inputs = tok(["A list of colors"], return_tensors="pt")
generate_ids=llm.generate(**model_inputs, max_new_tokens=5)
text=tok.batch_decode(generate_ids, skip_special_tokens=True)[0]
print(text)
```

```python
import torch
output_ids = model_inputs.input_ids[0].tolist()
for i in range(6):
  inputs = torch.tensor([output_ids], dtype=torch.long)
  outputs = llm(inputs)
  output_id = outputs.logits[:, -1].argmax()
  output_ids.append(output_id)
print(tok.decode(output_ids, skip_special_tokens=True))
```

```python
import gensim
import nltk
import numpy as np
nltk.download('word2vec_sample')
```

```python
word2vec_sample = str(nltk.data.find('models/word2vec_sample/pruned.word2vec.txt'))
model = gensim.models.KeyedVectors.load_word2vec_format(word2vec_sample,binary=False)
```

```python
king = model['king']
king[:10]
```

```python
man = model['man']
women = model['women']
```

```python
answer= king - man + women
```

```python
similarites = model.vectors.dot(answer)
indices = np.argsort(-similarites, axis=0)

for index in indices[:3]:
  word = model.index_to_key[index]
  if word not in ['king', 'man', 'women']:
    print(word)
    break
```

```python
model.most_similar(positive=['king','women'], negative=['man'],topn=1)
```

```python
model.most_similar(positive=['India','Paris'], negative=['Germany'],topn=1) #india-germany+paris
```

```python
model.most_similar(positive=['cat','tree'], negative=['bird'],topn=2)#cat-bird+tree
```
